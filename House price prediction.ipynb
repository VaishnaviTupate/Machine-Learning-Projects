# Visualize Data

```python
X = data.drop('price',axis='columns')
X
```

```python
Y = data.price
Y
```

# Training Dataset using Linear Regression

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
x=1265
LandAreainSqFt=[[x]]
PredictedmodelResult = model.predict(LandAreainSqFt)
print(PredictedmodelResult)
```

# Lets Check our Model is Right
Theory Calculation Y=m*X+b (m is coefficient & b is intercept) 
Coefficient-m

```python
m=model.coef_
print(m)
```

```python
b=model.intercept_
print(b)
```

# Y = mx + c
x is Independent variable - Input - Area

```python
y = m*x + b
print("the price of {0} square feet Land is: {1}". format (x,y[0]))
```
